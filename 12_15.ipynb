```python
import matplotlib as mpl
import numpy as np
import pandas as pd
import plotly.express as px

data = pd.read_csv("Electric_Vehicle_Population_Data.csv")

data = data.dropna()
data = data.drop_duplicates() # no duplicates

data.info() #because of privacy, the VIN numbers are fake; 20+ car registrations under same VIN number.
```

```python
County_count = data["County"].value_counts()
County_count
```

```python
data_king_county = data[data["County"] == "King"]
data_king_county["City"].value_counts()
```

```python
#how many EVs in each county
bar_plot_1 = data['County'].value_counts()

bar_plot_1.plot(
    kind = 'bar',
    title = 'Number of EVs in each County (WA)',
    xlabel = 'County',
    ylabel = 'Count',
    figsize = (8,10),
    color = "green"
)

bar_plot_1
```

```python
#most popular EV in Washington - make 0 value_counts()

bar_plot_2 = data['Make'].value_counts()
bar_plot_2 = bar_plot_2.head(10)


bar_plot_2.plot(
    kind = 'bar',
    title = "Top 10 EV Brands in WA",
    xlabel = "EV Brand",
    ylabel = "Count",
    figsize = (8,15)
    )

bar_plot_2.head(10)
```
